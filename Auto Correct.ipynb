```python
import re
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# the tiny corpus of text ! 
text = 'red pink pink blue blue yellow ORANGE BLUE BLUE PINK'.lower()
print(text)
print('string length : ',len(text))
```

```python
words = re.findall(r"\w+",text)
print(words)
```

```python
vocab = set(words)
print(vocab)
```

```python
count_a = dict()
for w in words:
  count_a[w] = count_a.get(w,0)+1
print(count_a)
```

```python
count_b = Counter(words)
print(count_b)
```

# Edits

```python
word = "dearz"
```

```python
splits = []
for i in range(len(word)+1):
  splits.append([word[:i],word[i:]])
print(splits)
```

```python
deletes = [L+R[1:] for L,R in splits if R]
edits = list(deletes)
edits
```

```python
vocab = ['dean','deer','dear','fries','and','coke']
edits = list(deletes)

print('vocab : ', vocab)
print('edits : ', edits)

candidates=[]

for w in edits:
    if w in vocab:
        candidates.append(w)

print('candidate words : ', candidates)
```
